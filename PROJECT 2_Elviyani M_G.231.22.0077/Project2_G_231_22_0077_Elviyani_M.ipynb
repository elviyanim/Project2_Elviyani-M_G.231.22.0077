```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
df = pd.read_csv("file_project2_1.csv")

x_col = 'Bulan'
y_col= ['1981-1990','1991-2000','2001-2010']
colors=['blue','red','green']
a=df.plot(x=x_col, y=y_col, kind='bar', color=colors, zorder=2)
plt.ylabel('Curah Hujan (mm)', fontsize=12, fontweight='bold')
plt.xlabel('Bulan', fontsize=12, fontweight='bold')

a.set_ylim(0, 680)
a.yaxis.set_major_locator(MultipleLocator(100))
a.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which='minor', color='#CCCCCC')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
df = pd.read_csv("file_project2_2.csv")

x_col = 'Waktu'
y_col= ['Asimilasi','Observasi','Non Asimilasi']
colors=['blue','yellow','red']
a=df.plot(x=x_col, y=y_col, kind='line', color=colors, marker='.')
plt.ylabel('Celcius', fontsize=12, fontweight='bold')
plt.xlabel('Waktu', fontsize=12, fontweight='bold')
plt.legend(loc='lower right',bbox_to_anchor=(1, 0),shadow=True)
a.set_ylim(20,27)
a.yaxis.set_major_locator(MultipleLocator(1))
a.grid(which='minor', color='#CCCCCC')
plt.grid(zorder=1, axis='y', linewidth='0.5', linestyle="--")


plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
